```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
im = Image.open("DTEEC_069895_2350_070528_2350_A01.ab.jpg")

print(im.format, im.size, im.mode) #the RGB value due to JPEG
```

```python
imgplot = plt.imshow(im)
```

```python
imdata = np.array(im, dtype=int)
#imgplot2.set_cmap('prism')
```

```python
imdatax = np.linspace(start=0, stop=im.size[1],num=im.size[1])
```

```python
toextract = 600 # where the matrix line reads
imdataRot = np.rot90(imdata)
imdataRot = imdataRot[600:1800,:]
imdataRotADDLINE = imdataRot.copy()
imdataRotADDLINE[toextract,:] = [100, 100, 100] #read matrix from row 600 (above) at 1st line of code
plt.figure(figsize=(17, 3))
imgplot2 = plt.imshow(imdataRotADDLINE)
plt.savefig('HW test.jpg', format='jpg', dpi=500)
```

```python
imdataRED = imdataRot[toextract,:][:,0]
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
# Plot and show our data
ax.plot(imdatax,imdataRED, color='r', label='red')
plt.xlim([250, 2200])
#plt.ylim([700, 1000])
plt.xlabel('Distance [AU]')
plt.ylabel('Height [AU]')
plt.title('Gully Monitoring and Topography Line Profile')
plt.savefig('profile line.jpg', format='jpg', dpi=300)
plt.show()
```
